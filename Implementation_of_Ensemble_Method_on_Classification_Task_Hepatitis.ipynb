```python
def prepare_dataset(df):
    # Mengambil fitur-fitur dari dataset ke dalam variabel X, dan target (class) ke dalam variabel y
    X = df.drop(['class'], axis=1)
    y = df.iloc[:, -1]

    # Membuat objek scaler untuk normalisasi data menggunakan MinMaxScaler dengan rentang (-1, 1)
    scaler = MinMaxScaler(feature_range=(-1, 1))
    # Melakukan normalisasi terhadap fitur-fitur dalam X
    X = scaler.fit_transform(X)

    # Membagi dataset menjadi data latih (train) dan data uji (test) dengan perbandingan 70:30
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

    # Mengembalikan data-data yang sudah dipersiapkan
    return X_train, X_test, y_train, y_test

def evaluation_metrics(y_test, y_pred):
    cm = confusion_matrix(y_test,y_pred)
    sns.set(rc={"figure.figsize":(8, 4)})
    heatmap = sns.heatmap(cm, annot=True,cmap="crest")

    print(classification_report(y_test,y_pred))
    print("Akurasi:",round(accuracy_score(y_test,y_pred) * 100 , 2),"%\n")

def clf_performance(classifier, model_name):
    print(model_name)
    print('Best Score: ' + str(classifier.best_score_))
    print('Best Parameters: ' + str(classifier.best_params_))
```

#### **Bagging**
Proses "Bagging Tuning" merujuk pada upaya untuk mengoptimalkan performa model Bagging melalui penyesuaian parameter-parameter tertentu. Bagging (Bootstrap Aggregating) adalah metode ensemble di mana sejumlah model paralel, seperti decision tree, dibangun dengan menggunakan subset acak dari data pelatihan. Proses tuning melibatkan penyetelan parameter-parameter algoritma Bagging untuk mencapai kinerja yang lebih baik.

```python
bagging_param = {
    'n_estimators' : [50,100,150,200], # Jumlah pohon dalam ensambel.
    'max_samples' : [5,10,50], # Ukuran maksimum subsample untuk setiap pohon.
    'estimator' : [DecisionTreeClassifier(random_state=1)], # Estimator yang digunakan sebagai model dasar (pada kasus ini, DecisionTreeClassifier dengan seed acak 1).
    "estimator__max_depth": [2, 4, 6], # Kedalaman maksimum dari setiap pohon dalam ensambel.
  }

bagging_tuning = GridSearchCV(bagging,
                         param_grid = bagging_param,
                         cv = 5, # Validasi silang dengan 5 lipatan
                         scoring='accuracy',
                         verbose = True,
                         n_jobs = -1)

best_bagging = bagging_tuning.fit(X_train,y_train)

clf_performance(best_bagging,'Bagging')
```

```python
bagging_param = {
    'n_estimators' : best_bagging.best_params_["n_estimators"],
    'max_samples' : best_bagging.best_params_["max_samples"],
    'estimator' : best_bagging.best_params_["estimator"],
  }

# mengimplementasikan best parameter kembali kedalam model

bagging_tuning = BaggingClassifier(**bagging_param)
bagging_tuning.fit(X_train, y_train)
```

```python
bagging_tuning_y_pred = bagging_tuning.predict(X_test)
model_comparison['Bagging Tuning'] = f"{round(accuracy_score(y_test,bagging_tuning_y_pred) * 100 , 2)}%"
evaluation_metrics(y_test, bagging_tuning_y_pred)
```

#### **Boosting**
Proses "Boosting Tuning" merujuk pada upaya untuk mengoptimalkan performa model Boosting melalui penyetelan parameter-parameter tertentu. Boosting adalah metode ensemble di mana model-model lemah (weak learners), seperti decision tree yang dangkal, dibangun secara berurutan, dan setiap model berusaha untuk memperbaiki kesalahan prediksi model sebelumnya. Proses tuning melibatkan penyesuaian parameter-parameter algoritma Boosting untuk mencapai kinerja yang lebih baik.

```python
boosting_param = {
        'n_estimators' : [10,20,30,40,50], # Jumlah pohon dalam ensambel.
        'learning_rate' : [0.01, 0.1, 1], #Tingkat pembelajaran yang mengontrol seberapa besar kontribusi dari setiap model.
        'estimator' : [DecisionTreeClassifier(random_state=1)], #Estimator yang digunakan sebagai model dasar
        "estimator__max_depth": [2, 4, 6], #Kedalaman maksimum dari setiap model dalam ensambel.
        }

boosting_tuning = GridSearchCV(boosting,
                         param_grid = boosting_param,
                         cv = 5,
                         scoring='accuracy',
                         verbose = True,
                         n_jobs = -1)

best_boosting = boosting_tuning.fit(X_train,y_train)

clf_performance(best_boosting,'Boosting')
```

```python
boosting_param = {
    'estimator': best_boosting.best_params_["estimator"],
    'learning_rate':best_boosting.best_params_["learning_rate"],
    'n_estimators': best_boosting.best_params_["n_estimators"]
    }

# mengimplementasikan best parameter kembali kedalam model
boosting_tuning = AdaBoostClassifier(**boosting_param)
boosting_tuning.fit(X_train, y_train)
```

```python
boosting_tuning_y_pred = boosting_tuning.predict(X_test)
model_comparison['Boosting Tuning'] = f"{round(accuracy_score(y_test,boosting_tuning_y_pred) * 100 , 2)}%"
evaluation_metrics(y_test, boosting_tuning_y_pred)
```
